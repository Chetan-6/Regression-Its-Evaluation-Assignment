Que 1. What is Simple Linear Regression ?

Ans - ‚≠ê Simple Linear Regression (SLR)
- It is a technique used to predict a continuous (numeric) value using one independent variable (feature).

üìå Formula:

Y=a+bX

Where:
- Y = Target / value to be predicted
- X = Independent variable (input feature)
- a = Intercept (value of Y when X = 0)
- b = Slope (how much Y changes when X increases by 1 unit)

‚≠ê Example:

Predicting salary (Y) based on years of experience (X).

If the regression line is:

Salary=30,000+4,000 * Experience

Then:
- For 2 years experience ‚Üí Salary = 38,000
- For 5 years experience ‚Üí Salary = 50,000

‚≠ê Goal of Simple Linear Regression:
  
  To find the best-fit straight line through the data that:
- Minimizes the error between actual and predicted values
- Helps understand the relationship between X and Y

‚≠ê Where is it used?

- Predicting sales based on advertising spend
- Predicting weight from height
- Predicting bike/car price from engine power
- Predicting marks from hours studied

‚≠ê Key Terms

| Term                         | Meaning                    |
| ---------------------------- | -------------------------- |
| **Independent Variable (X)** | Input feature              |
| **Dependent Variable (Y)**   | Output we want to predict  |
| **Slope (b)**                | Rate of change             |
| **Intercept (a)**            | Base value                 |
| **Residual**                 | Error = Actual ‚Äì Predicted |


Que 2. What are the key assumptions of Simple Linear Regression ?

Ans -

‚úÖ 1. Linearity:

The relationship between the independent variable (X) and the dependent variable (y) must be linear.
If the relationship is curved or non-linear, linear regression won‚Äôt work well.

‚úÖ 2. Independence of Errors:

The residuals (errors) should be independent of each other.
This means one error should not depend on another ‚Äî no pattern between errors.

‚úÖ 3. Homoscedasticity (Constant Variance of Errors):

The variance of residuals should be constant for all values of X.
If the spread of residuals increases/decreases, that's heteroscedasticity, which is a problem.

‚úÖ 4. Normality of Errors:

Residuals should follow a normal distribution, especially important for hypothesis testing and confidence intervals.

‚úÖ 5. No or Minimal Outliers:

Extreme outliers can strongly affect the regression line, so data should be checked for them.

üö´ (Not required in Simple Linear Regression):
- X does not need to be normally distributed
- y does not need to be normally distributed
- X does not need to be independent from y (but they should be related)




Que 3. What is heteroscedasticity, and why is it important to address in regression models ?

Ans -

‚úÖ What is Heteroscedasticity?

In regression, heteroscedasticity means:
  - The variance of the errors (residuals) is not constant across all levels of the independent variable(s).

In simple words:
- When you plot residuals, they should be evenly spread.
- If they fan out, narrow, or change pattern ‚Üí this is heteroscedasticity.

Example:

Suppose you predict income based on experience.

If low-experience people have small errors but high-experience people have huge errors ‚Üí heteroscedasticity exists.

üéØ Why is it important to address?

Heteroscedasticity does NOT change the regression coefficients,
but it does affect the reliability of your model.

üîπ 1. It makes standard errors wrong

When variance is not constant, standard errors become:
- too small ‚Üí false significance
- too large ‚Üí missing significant predictors

This leads to incorrect p-values and confidence intervals.

üîπ 2. It violates a key assumption of Linear Regression

One assumption is:
  - Errors must have constant variance (homoscedasticity).

If this assumption fails ‚Üí results are not trustworthy.

üîπ 3. It reduces model performance

The model becomes:
- Less accurate
- Less stable
- Bad at prediction for certain ranges of data


üîπ 4. It affects hypothesis testing

t-tests, F-tests, and overall model reliability depend on the assumption of constant variance.

So heteroscedasticity can cause wrong conclusions.


‚úîÔ∏è How to detect heteroscedasticity?

1. Residual vs Fitted plot

Look for cone-shaped or curved patterns.

2. Breusch‚ÄìPagan test

3. Goldfeld‚ÄìQuandt test



‚úîÔ∏è How to fix it?

‚úîÔ∏è 1. Use log or sqrt transformation

If y is skewed.

‚úîÔ∏è 2. Weighted least squares (WLS)

‚úîÔ∏è 3. Use robust standard errors

(e.g., Huber-White SEs)

üìå Simple Summary

| Term                   | Meaning                                       |
| ---------------------- | --------------------------------------------- |
| **Heteroscedasticity** | Error variance not constant                   |
| **Problem?**           | Wrong SEs ‚Üí wrong p-values ‚Üí unreliable model |
| **Detect?**            | Residual plots, BP test                       |
| **Fix?**               | Transform, WLS, robust errors                 |




Que 4. What is Multiple Linear Regression ?

Ans -

Multiple Linear Regression (MLR) is a supervised machine learning and statistical technique used to model the relationship between one dependent variable and two or more independent variables.

‚úÖ Definition:

Multiple Linear Regression predicts a continuous output using the formula:

Y=b0‚Äã+b1‚ÄãX1‚Äã+b2‚ÄãX2‚Äã+b3‚ÄãX3‚Äã+‚Ä¶+bn‚ÄãXn‚Äã


Where:
- Y = target variable (what you want to predict)
- X‚ÇÅ, X‚ÇÇ, ‚Ä¶, X‚Çô = input features
- b‚ÇÄ = intercept
- b‚ÇÅ, b‚ÇÇ, ‚Ä¶, b‚Çô = coefficients (how much each feature contributes to prediction)

‚úÖ Example

Predicting a house price using:
- Size of house (sq ft)
- Number of bedrooms
- Location score
- Age of the building

MLR finds how each feature influences the price.

‚úÖ Where is MLR used?
- Price prediction (bikes, cars, houses)
- Salary prediction
- Sales forecasting
- Medical measurements (predicting blood pressure, etc.)
- Marketing (effect of ads on sales)

‚úÖ Key Assumptions
1. Linearity ‚Äì relationship between variables is linear
2. No multicollinearity ‚Äì features shouldn‚Äôt be highly correlated
3. Homoscedasticity ‚Äì constant variance of errors
4. Normality of residuals
5. Independence of errors

‚úÖ Advantages
- Simple and interpretable
- Shows which features are important
- Fast to train

‚ùå Disadvantages
- Doesn‚Äôt work well with non-linear data
- Sensitive to outliers
- Bad performance if features are correlated



Que 5. What is polynomial regression, and how does it differ from linear regression ?

Ans -

‚úÖ What is Polynomial Regression?

Polynomial Regression is a type of regression analysis where the relationship between the independent variable (X) and dependent variable (y) is modeled as an nth-degree polynomial.

Example:

y=Œ≤0‚Äã+Œ≤1‚Äãx+Œ≤2‚Äãx2+Œ≤3‚Äãx3+‚ãØ+Œ≤n‚Äãxn

It can model curved or non-linear patterns in the data.

‚úÖ How Does It Differ From Linear Regression?

| Feature                     | Linear Regression           | Polynomial Regression                           |
| --------------------------- | --------------------------- | ----------------------------------------------- |
| **Type of relationship**    | Straight line               | Curved line                                     |
| **Equation form**           | ( y = \beta_0 + \beta_1 x ) | ( y = \beta_0 + \beta_1 x + \beta_2 x^2 + ... ) |
| **Captures non-linearity?** | ‚ùå No                        | ‚úîÔ∏è Yes                                          |
| **Model complexity**        | Low                         | Higher                                          |
| **Overfitting risk**        | Low                         | High (if degree is large)                       |


‚úÖ Key Point

Polynomial regression is still linear in terms of coefficients (Œ≤‚Äôs).

Only X is transformed, not the model type.

So it is solved using the same method as linear regression (least squares).



üß† When to Use Polynomial Regression?

Use it when the data shows:
- curved relationship
- trends that increase/decrease and then reverse
- non-linear patterns that are not captured by a straight line




Que 6. Implement a Python program to fit a Simple Linear Regression model to
the following sample data:

‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.1, 4.3, 6.1, 7.9, 10.2]

Plot the regression line over the data points.

Ans -

The fitted simple linear regression line is approximately [y =¬†1.98x¬†+¬†0.18]



```python
#Code:

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Fit Simple Linear Regression
model = LinearRegression()
model.fit(X, Y)

slope = model.coef_[0]
intercept = model.intercept_

print(f"Slope (m): {slope}")
print(f"Intercept (c): {intercept}")
print(f"Equation of line: y = {slope:.2f}x + {intercept:.2f}")

# Predictions for plotting
Y_pred = model.predict(X)

# Plot data points and regression line
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Regression line')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Simple Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Model fitting Code:

import numpy as np
from sklearn.linear_model import LinearRegression

# Data
area = np.array([1200, 1500, 1800, 2000])
rooms = np.array([2, 3, 3, 4])
price = np.array([250000, 300000, 320000, 370000])

# Feature matrix (Area, Rooms)
X = np.column_stack((area, rooms))
Y = price

# Fit Multiple Linear Regression
model = LinearRegression()
model.fit(X, Y)

# Parameters
coef_area, coef_rooms = model.coef_
intercept = model.intercept_

print(f"Coefficient for Area: {coef_area}")
print(f"Coefficient for Rooms: {coef_rooms}")
print(f"Intercept: {intercept}")
print(f"Equation: Price = {coef_area:.2f}*Area + {coef_rooms:.2f}*Rooms + {intercept:.2f}")
```

Que 8. Implement polynomial regression on the following data:

‚óè X = [1, 2, 3, 4, 5]
‚óè Y = [2.2, 4.8, 7.5, 11.2, 14.7]

Fit a 2nd-degree polynomial and plot the resulting curve.

Ans -

A 2nd-degree polynomial fit for the given data is approximately
[y = 0.20x^2 +¬†1.94x¬†+¬†0.06].


```python
# Code:

import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression

# Data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features (degree 2)
poly = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly.fit_transform(X)

# Fit Polynomial Regression (using Linear Regression on polynomial features)
model = LinearRegression()
model.fit(X_poly, Y)

# Get coefficients
coef = model.coef_
intercept = model.intercept_
print("Intercept:", intercept)
print("Coefficients:", coef)
print(f"Equation: y = {coef[1]:.2f}x^2 + {coef[0]:.2f}x + {intercept:.2f}")

# Generate smooth curve for plotting
X_plot = np.linspace(1, 5, 100).reshape(-1, 1)
X_plot_poly = poly.transform(X_plot)
Y_plot = model.predict(X_plot_poly)

# Plot original data and polynomial curve
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X_plot, Y_plot, color='red', label='2nd-degree fit')

plt.xlabel('X')
plt.ylabel('Y')
plt.title('Polynomial Regression (Degree 2)')
plt.legend()
plt.grid(True)
plt.show()
```

This program fits a quadratic polynomial to the data and plots the resulting smooth curve over the original¬†points.

Que 9.  Create a residuals plot for a regression model trained on this data:


‚óè X = [10, 20, 30, 40, 50]
‚óè Y = [15, 35, 40, 50, 65]

Assess heteroscedasticity by examining the spread of residuals.

Ans -

The residuals show no clear increasing or decreasing spread with X, so strong heteroscedasticity is not evident on this¬†small¬†sample.




```python
# Code:

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Fit linear regression
model = LinearRegression()
model.fit(X, Y)

# Predictions and residuals
Y_pred = model.predict(X)
residuals = Y - Y_pred

print("Predictions:", Y_pred)
print("Residuals:", residuals)

# Residual plot
plt.scatter(X, residuals, color='blue')
plt.axhline(y=0, color='red', linestyle='--')

plt.xlabel('X')
plt.ylabel('Residuals (Y - ≈∂)')
plt.title('Residual Plot')
plt.grid(True)
plt.show()
```

Interpreting heteroscedasticity:
- The residuals for this fit are approximately: [[-3.0,\ 5.5,\ -1.0,\ -2.5,\ 1.0]] for increasing X values.
- With only five points and a mixed pattern of positive and negative residuals without a clear funnel shape, strong evidence of heteroscedasticity cannot be concluded¬†from¬†this¬†plot.

Que 10.  Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.

Ans -

‚úÖ How to Handle Heteroscedasticity & Multicollinearity in a House Price Prediction Model:

As a data scientist building a house-price prediction model, detecting heteroscedasticity and multicollinearity signals that the linear regression assumptions are violated. To ensure a robust and reliable model, I would follow these steps:

üü¶ 1. Addressing Heteroscedasticity (Non-constant variance in errors)

üìå Step 1: Visual diagnosis
- Plot residuals vs. fitted values.
- Look for patterns like a funnel shape, increasing spread, etc.

üìå Step 2: Statistical tests
- Breusch‚ÄìPagan test
- White test

üìå Step 3: Apply corrective techniques

 üîß Option A: Log Transformation

If price variance increases with area or rooms:
- Transform Price ‚Üí log(Price)
- Sometimes transform Area ‚Üí log(Area)

Effect: Stabilizes variance and linearizes the relationship.

  üîß Option B: Weighted Least Squares (WLS)

- Assign lower weights to observations with large variance.
- Useful when form of heteroscedasticity is known.

  üîß Option C: Robust Standard Errors
- Use Heteroscedasticity-consistent (HC) standard errors
- Useful when heteroscedasticity is mild.


üü¶ 2. Addressing Multicollinearity (High correlation among predictors)

üìå Step 1: Identify multicollinearity
- Compute correlation matrix
- Check Variance Inflation Factor (VIF)
  - VIF > 5 (or 10) indicates multicollinearity.

üìå Step 2: Apply corrective techniques

  üîß Option A: Drop one of the correlated features

Example:
- Area & rooms might be highly correlated
  
  ‚Üí Keep the stronger predictor; drop the weaker one.

üîß Option B: Combine correlated variables
- Create new features like Area per room, Total living area, etc.

üîß Option C: Regularization models

  Switch to models that naturally handle multicollinearity:
- Ridge Regression ‚Üí shrinks coefficients
- Lasso Regression ‚Üí can remove irrelevant features
- Elastic Net ‚Üí mix of Ridge + Lasso


üîß Option D: PCA (Dimensionality Reduction)
- Convert correlated variables into uncorrelated principal components.

üü¶ 3. Build the Final Robust Model

After fixing the issues, I would:
1. Recalculate VIF and recheck residual plots
2. Refit regression with modified features
3. Compare models using:
  - Adjusted R¬≤
  - RMSE
  - AIC / BIC
4. Use cross-validation (k-fold) to ensure generalizability
5. Deploy the best-performing, stable model

üü¢ Final Summary (Short version)

To ensure a robust house-price model:
- Fix heteroscedasticity using log transformations, WLS, or robust SEs
- Fix multicollinearity using VIF-based feature removal, combining features, or regularization (Ridge/Lasso)
- Re-evaluate the model with diagnostics and cross-validation
